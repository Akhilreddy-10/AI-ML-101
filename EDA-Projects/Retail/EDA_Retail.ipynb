# Exploratory Data Analysis 

**Data Visualization**

```python
correlation = df[['Sales','Discount','Quantity','Profit']].corr()
sns.heatmap(correlation,annot=True,cmap='Greens')
plt.title('Correlation')
```

Inferences:
-  No such strong correlation.
-  Discount and Profit are Negatively Correlated.
-  Profit and Sales are Positively Correlated.

```python
Profit_plot = sns.catplot(x='Sub-Category', y='Profit', data=df, kind='bar', palette='Spectral')
Profit_plot.fig.suptitle('Box Plot')
plt.xticks(rotation=90)
plt.show()
```

Inferences:
-  Sub-Category Machines has suffered the highest Loss.
-  Copiers has highest profit with no Loss.

```python
figsize=(15,10)
sns.pairplot(df,hue='Sub-Category')
plt.show
```

Inference: 
-  Our data is not normal.
-  It has some amount of outliers too.

We will explore about this outliers using boxplots.

```python
sns.scatterplot(x='Discount', y='Profit', hue='Category', data=df)
plt.show()
```

Inferences
-  Category Technology shows maximum Profit with no Discount.
-  Category Furniture has very minor difference in profits earned for different discount rates.
-  Category Office Supplies shows maximum profit at no Discount.

### Sales by Segment

```python
sales_by_segment = df.groupby('Segment').size()
sales_by_segment = sales_by_segment.plot(title='Sales by Segment', color='maroon')
sales_by_segment.set_xlabel('Segment')
sales_by_segment.set_ylabel('Sales')
```

```python
sns.set_style('whitegrid')
df_sales = df.groupby(['Segment', 'Sub-Category'])['Sales'].sum().reset_index()
g = sns.catplot(x='Sub-Category', y='Sales', data=df_sales, kind='bar', palette='bright', col='Segment')
g.set_xticklabels(rotation=90)
g.set_axis_labels("Sub-Category", "Sales")
plt.subplots_adjust(top=0.9)
plt.suptitle('Sales From Every Segment Of United States of Whole Data')
plt.show()
```

```python
df_segment= df.groupby(['Segment'])[['Sales','Discount','Profit']].mean()
df_segment
```

Inference:
-  Consumer Segment has most purchased Sub-Categories.
-  Home Office has least purchased Sub-Categories.
-  Fasteners have lowest sales compared to all sub-categories in each segment.
-  Chairs and Phones sales are highest compared to all sub-categories in each segment.

### Sales by Region

```python
sales_by_region = df.groupby('Region').size()
plot_by_region = sales_by_region.plot(title='Sales by Region',xticks=(1,2,3,4,),color='olivedrab')
plot_by_region.set_xlabel('Region')
plot_by_region.set_ylabel('Sales')
```

```python
sns.set_style('whitegrid')
df_sales = df.groupby(['Region', 'Sub-Category'])['Sales'].sum().reset_index()
g = sns.catplot(x='Sub-Category', y='Sales', data=df, kind='bar', palette='bright', col='Region')
g.set_xticklabels(rotation=90)
g.set_axis_labels("Sub-Category", "Sales")
plt.subplots_adjust(top=0.9)
plt.suptitle('Sales From Every Region Of United States of Whole Data')
plt.show()
```

Inferences:

-  Sales are highest in west region of the country.
-  Machines are most sold product in south region whereas copiers are most sold product in other regions.

### Sales by Ship-Mode

```python
sales_by_shipmode = df.groupby('Ship Mode').size()
sales_by_shipmode = sales_by_shipmode.plot(title='Sales by Ship Mode',color='darkgreen')
sales_by_shipmode.set_xlabel('Ship Mode')
sales_by_shipmode.set_ylabel('Sales')
```

```python
sns.set_style('whitegrid')
df_sales = df.groupby(['Ship Mode', 'Sub-Category'])['Sales'].sum().reset_index()
g = sns.catplot(x='Sub-Category', y='Sales', data=df, kind='bar', palette='bright', col='Ship Mode')
g.set_xticklabels(rotation=90)
g.set_axis_labels("Sub-Category", "Sales")
plt.subplots_adjust(top=0.9)
plt.suptitle('Sales From Every Ship Mode Of United States of Whole Data')
plt.show()
```

Inferences:

-  Sales are highest in Standard Class.
-  Sales are lowest in Same day.
-  Sales of Machines are highest in Second Class and Same Day.
-  Sales of Copiers are highest in Standard and First Class.

### Deal Analysis of State

```python
df2 = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
df2[:].plot.bar(color=['hotpink','goldenrod'],figsize=(15,10))
plt.title('Profit/Loss and Sales Across States')
plt.xlabel('States')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

Inferences:

-  Countries where there is loss on sales are:
   1. Texas
   2. Pennsylvania
   3. Florida
   4. Ilinois
   5. Ohio
   6. North Carolina
   7. Arizona
   8. Colorado
   9. Tennessee
   
 <br>

## Conclusion

**Business Manager should feel happy to know that SampleSuperstore is already in Profit.**

But still there are some areas where we can work over to maximize the profit.

Let's look into it.

-  Discounts of 20%-30% shows more profit for most of the products, so there is no need of giving discounts more than 50% on most of the products.
- Different situations which we seen through data analysis could be looked over in the states where the is loss and decisions should be taken accordingly.
- Sales could be increased by a certain rate where there is no profit to check over, otherwise company should stop it's sales over those states.
-  Instead of giving heavy discounts for clearance sales, sales could be increased in non-profit states.
-  On an average more sales means more profit.
